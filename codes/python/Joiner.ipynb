change the ride id column to string

```python
q2_2019_nw= q2_2019.astype({'ride_id': 'str'})
q3_2019_nw= q3_2019.astype({'ride_id': 'str'})
q4_2019_nw= q4_2019.astype({'ride_id': 'str'})
```

latitude and longitude data are only q1_2020 dataset, thus we drop the columns

```python
q1_2020.drop(columns=["start_lat","start_lng","end_lat","end_lng"],inplace=True)
```

Gender, birthyear columns are no longer used in q1_2020 thus we drop the columns from the other datasets

```python
q2_2019_nw.drop(columns=["Member Gender",	"05 - Member Details Member Birthday Year"],inplace=True)
q3_2019_nw.drop(columns=["gender","birthyear"],inplace=True)
q4_2019_nw.drop(columns=["gender","birthyear"],inplace=True)
```

drpo the trip duration columns from the other datasets than q1_2020 as it is not part of q1_2020

```python
q2_2019_nw.drop(columns=["01 - Rental Details Duration In Seconds Uncapped"],inplace=True)
q3_2019_nw.drop(columns=["tripduration"],inplace=True)
q4_2019_nw.drop(columns=["tripduration"],inplace=True)
```

```python
q3_2019_nw.rename(columns={'usertype':'member_casual'}, inplace=True)
q4_2019_nw.rename(columns={'usertype':'member_casual'}, inplace=True)
```

Check all the columns for consistency

```python
q1_2020.drop(columns=["rideable_type"],inplace=True)
q2_2019_nw.drop(columns=["01 - Rental Details Bike ID"],inplace=True)
q3_2019_nw.drop(columns=["bikeid"],inplace=True)
q4_2019_nw.drop(columns=["bikeid"],inplace=True)
```

```python
colswap=['ride_id', 'started_at', 'ended_at', 'start_station_name', 'end_station_name', 'start_station_id','end_station_id','member_casual']
```

Rearrange the columns to align columns all the datasets

```python
q1_2020_nw= q1_2020.reindex(columns=colswap)
q2_2019_nw= q2_2019_nw.reindex(columns=colswap)
q3_2019_nw= q3_2019_nw.reindex(columns=colswap)
q4_2019_nw= q4_2019_nw.reindex(columns=colswap)
```

Join the all the dataset

```python
last_year_full_data=pd.concat([q2_2019_nw,q3_2019_nw,q4_2019_nw,q1_2020_nw])
```

```python
last_year_full_data
```

Rename subscriber as members and customer as casual to align with member and casual categorization of q1_2020 dataset

```python
last_year_full_data['member_casual'] =last_year_full_data['member_casual'].replace(['Subscriber'],'member')
```

```python
last_year_full_data['member_casual'] =last_year_full_data['member_casual'].replace(['Customer'],'casual')
```
