```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
from google.colab import drive
drive.mount('/content/drive') 
```

```python
df=pd.read_csv(r"/content/drive/MyDrive/DA_portfolio/Divvy_Trips_cleaner.csv",index_col=0)
```

```python
df.head()
```

```python
drop=df.loc[(df["start_station_name"]=="HQ QR") & (df["end_station_name"]=="HQ QR") & (df["tripduration_sec"]<5)]
```

```python
df2=df.drop(drop.index)
```

```python
day_replace= { 0 : 'Monday', 1 : 'Tuesday', 2 : 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'} 
```

```python
df2['day of the week'] =df2['day of the week'].replace(day_replace)
```

```python
df2
```

```python
df2.loc[(df2["start_station_name"]=="HQ QR") & (df2["end_station_name"]=="HQ QR") & (df2["tripduration_sec"]<5)]
```

```python
pv_table1=pd.pivot_table(df2, values='tripduration_sec', index=["year","month"],
                    columns=['member_casual'], aggfunc=[np.mean,np.median,stats.mode,np.min,np.max,np.sum,np.count_nonzero,np.std])
```

```python
pv_table2=pd.pivot_table(df2, values='tripduration_sec', index=["day of the week"],
                    columns=['member_casual'], aggfunc=[np.mean,np.median,stats.mode,np.min,np.max,np.sum,np.count_nonzero,np.std])
```
